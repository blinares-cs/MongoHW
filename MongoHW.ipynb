### Exercise 1: Basic Searching and Filtering

```python
# Find the first movie that has the genre "Action"
action_movie = collection.find_one({"genres": "Action"})

# Print the first movie
pprint.pprint(action_movie)
```

```python
# Find all movies released after the year 2000 (Return the first 5 results)
movies_after_2000 = collection.find({"year": {"$gt": 2000}}).limit(5)

# Print first 5 results
for movie in movies_after_2000:
    pprint.pprint(movie)
```

### Exercise 2: Sorting Results

```python
# Find all movies where the genre is "Comedy" and sort them by IMDb rating in descending order (Return the first 5 results)
sorted_comedy_movies = collection.find({"genres": "Comedy"}).sort("imdb.rating", -1).limit(5)

# Print the first 5 results
for movie in sorted_comedy_movies:
    pprint.pprint(movie)
```

```python
# Find all movies where the genre is "Drama" and sort them by release year in ascending order (Return the first 5 results)
sorted_drama_movies = collection.find({"genres": "Drama"}).sort("year", 1).limit(5)

# Print the first 5 results
for movie in sorted_drama_movies:
    pprint.pprint(movie)
```

### Exercise 3: Aggregation Pipeline

```python
# Calculate the average IMDb rating for each genre (Return the top 5 genres)
avg_rating_by_genre = collection.aggregate([
    {"$unwind": "$genres"},
    {"$group": {"_id": "$genres", "avg_rating": {"$avg": "$imdb.rating"}}},
    {"$sort": {"avg_rating": -1}},
    {"$limit": 5}
])


# Print the top 5 genres
for genre in avg_rating_by_genre:
    pprint.pprint(genre)
```

```python
# Find the top 5 directors by the average IMDb rating of their movies
top_directors = collection.aggregate([
    {"$group": {"_id": "$directors", "avg_rating": {"$avg": "$imdb.rating"}}},
    {"$sort": {"avg_rating": -1}},
    {"$limit": 5}
])

# Print the top 5 directors
for director in top_directors:
    pprint.pprint(director)
```

```python
# Calculate the total number of movies released in each year (Sort the results by the year)
movies_per_year = collection.aggregate([
    {"$group": {"_id": "$year", "total_movies": {"$sum": 1}}},
    {"$sort": {"_id": 1}}
])


# Print the movies
for year in movies_per_year:
    pprint.pprint(year)
```

### Exercise 4: Updating and Deleting Documents

```python
# Update the IMDb rating of a movie with the title "The Godfather" to 9.5
collection.update_one({"title": "The Godfather"}, {"$set": {"imdb.rating": 9.5}})
```

```python
# Update all movies where the genre is "Horror" and set their IMDb rating to 6.0 if it is currently null
collection.update_many({"genres": "Horror", "imdb.rating": {"$exists": False}}, {"$set": {"imdb.rating": 6.0}})
```

```python
# Delete all movies that were released before the year 1950
collection.delete_many({"year": {"$lt": 1950}})
```

### Exercise 5: Text Search

```python
# Drop existing index
collection.drop_index("cast_text_fullplot_text_genres_text_title_text")

# Create a new text index only on the 'title' field
collection.create_index([("title", "text")])

# Find movies with 'love' in the title
love_movies = collection.find({"$text": {"$search": "love"}})


# Print the movies
for movie in love_movies:
    pprint.pprint(movie)
```

```python
# Drop the existing 'title_text' index
collection.drop_index("title_text")

# Create a new text index on 'title' and 'plot' fields
collection.create_index([("title", "text"), ("plot", "text")])

# Text search across title and plot, sorted by IMDb rating
war_movies = collection.find({"$text": {"$search": "war"}}).sort("imdb.rating", -1).limit(5)

# Print the movies
for movie in war_movies:
    pprint.pprint(movie)
```
